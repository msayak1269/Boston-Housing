```python
#corr() fn calculate correlation b/w every column of data
corr=df.corr('pearson')

#take absolute values of correlation
corrs=[abs(corr[attr]['target']) for attr in list(features)]

#make a list of pairs(corr,features)
l=list(zip(corrs,list(features)))

#sort the list of pairs in descending order
#with the correlation value with the key of sorting
l.sort(key=lambda x:x[0],reverse=True)

#unzip pairs of two list
#zip(*l) take a list that look like [[a,b,c],[d,e,f],[g,h,i]]
#and returs [[a,d,g],[b,e,h],[c,f,i]]
corrs,labels=list(zip((*l)))

#plot the correlation w.r.t the target as a bar graph
index=np.arange(len(labels))
plt.figure(figsize=(15,5))
plt.bar(index,corrs,width=0.5)
plt.xlabel('Attributes')
plt.ylabel('Correlation with target')
plt.xticks(index,labels)
plt.show()
```

```python
#normalization of data
X=df['LSTAT'].values
Y=df['target'].values
print(Y[:5])
x_scaler=MinMaxScaler()
X=x_scaler.fit_transform(X.reshape(-1,1))
X=X[:,-1]
y_scaler=MinMaxScaler()
Y=y_scaler.fit_transform(Y.reshape(-1,1))
Y=Y[:,-1]
print(Y[:5])
```

```python
def error(m,x,c,t):
    n=x.size
    e=sum(((m*x+c)-t)**2)
    return e*1/( 2*n)
```

```python
#split the data set into training and testing data
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2)
```

```python
def update(m,x,c,t,learning_rate):
    grad_m=sum(2*((m*x+c)-t)*x)
    grad_c=sum(2*((m*x+c)-t))
    m=m-grad_m*learning_rate
    c=c-grad_c*learning_rate
    return m,c
```

```python
def gradient_descent(inti_m,inti_c,x,t,learnig_rate,iterations,error_threshold):
    m=inti_m
    c=inti_c
    error_values=list()
    mc_values=list()
    for i in range(iterations):
        e=error(m,x,c,t)
        if e<error_threshold:
            print("Error value less than error threshold.stopping gradient descent")
            break
        error_values.append(e)
        m,c=update(m,x,c,t,learning_rate)
        mc_values.append((m,c))
    return m,c,error_values,mc_values
```

```python
%%time
inti_m=0.9
inti_c=0
iterations=250
learning_rate=0.001
error_threshold=0.001

m,c,error_values,mc_values=gradient_descent(inti_m,inti_c,xtrain,ytrain,learning_rate,iterations,error_threshold)
```

```python
mc_value_anim=mc_values[0:250:5]
```

```python
fig,ax=plt.subplots()
ln,=plt.plot([],[],'ro-',animated=True)

def init():
    plt.scatter(xtest,ytest,color='g')
    ax.set_xlim(0,1.0)
    ax.set_ylim(0,1.0)
    return ln,
def update_frame(frame_number):
    m,c=mc_value_anim[frame_number]
    x1,y1=-0.5,m*-.5+c
    x2,y2=1.5,m*1.5+c
    ln.set_data([x1,x2],[y1,y2])
    return ln,
anim=FuncAnimation(fig,update_frame,frames=range(len(mc_value_anim)),init_func=init,blit=True)
HTML(anim.to_html5_video())
#plt.show()
```
